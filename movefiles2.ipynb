```python
import os
import csv
import shutil
import glob
import math 
import re

def movefiles(directory): 
    
    with open('objectlist.csv', mode='r') as inp:                              # creates a dictionary of objectID as the keys from 'objectlist.csv' (and empty values)          
        reader = csv.reader(inp)
        objiddict = {(str(rows[1])):('') for rows in reader}
                                  
    dirwalker = os.walk(directory)
    tbafolders = (next(dirwalker)[1])
    newpathdict = {}                                                           # dictionary where FINAL paths are stored                                                    
    tbapaths = {}                                                              # dictionary where all paths in 'to be added' are stored
    pathstbmdict = {}                                                          # dictionary where paths whose object ID (from objectlist.csv) is in a path from 'to be added'
    
    for path,directory,file in dirwalker:                                      # walks through directory and stores picture paths into tbapaths dictionary
        for pic in file: 
            newpaths = os.path.join(path,pic)
            tbapaths[newpaths] = ('')
    for tbapaths in tbapaths:                                                  # if objectid in the paths from 'to be added' it is stored in pathstbmdict
        for ids in objiddict: 
            if ids in tbapaths and len(ids) >= 5: 
                pathstbmdict[tbapaths] = ('')
                
    for keys in pathstbmdict:                                                  # keys are the paths of pictures that need to be moved
        if 'gems' in keys: 
            i = keys.index('s\\')                                              # splits string in order to get the objectID
            b = keys[i+1:] 
            
            if b[1] == 'G':
                idnum = b[1:8]
                foldernum = str(b[4]) + 'xxx'
                mainpath = '\\\\' + 'rcstore.rc.fas.harvard.edu\geology\TMS Media\Specimen\Gemstones' + '\\' + (foldernum) 
                if mainpath not in newpathdict.values(): 
                    newpathdict[idnum] = mainpath
                finaldest = shutil.move(keys,mainpath)                
            if b[1] == '2':
                idnum = (b[1:8]).strip('.')                
                foldernum = (str(b[1:5])) + '.' + 'xx'
                mainpath = '\\\\' + 'rcstore.rc.fas.harvard.edu\geology\TMS Media\Specimen\Gemstones' + '\\' + (foldernum) 
                if mainpath not in newpathdict.values(): 
                    newpathdict[idnum] = mainpath
                finaldest = shutil.move(keys,mainpath)
                
        if 'documents' in keys:
            i = keys.index('s\\')
            b = keys[i+1:]

            if b[1] == 'g':
                idnum = b[1:5]
                foldernum = str(b[3:])
                mainpath = '\\\\' + 'rcstore.rc.fas.harvard.edu\geology\TMS Media\Specimen\Accession and Object Records\Gem' + '\\' + (foldernum) 
                if mainpath not in newpathdict.values():
                    newpathdict[keys] = mainpath
                finaldest = shutil.move(newpaths,mainpath)
                    
            if b[1] != '2':
                image = re.findall("\d+",b)[0]
                if len(image) = 4: 
                    idnum = ((int(image))/10)
                    foldernum = math.floor(idnum)
                    foldernum = str(idnum) + '0'
                    mainpath = '\\\\' + 'rcstore.rc.fas.harvard.edu\geology\TMS Media\Archives\Accession and Object Records' + '\\' + (idnum)
                    if mainpath not in newpathdict.values(): 
                        newpathdict[idnum] = mainpath
                    finaldest = shutil.move(newpaths,mainpath)
                else:
                    idnum = ((int(image))/1000)
                    foldernum = math.floor(idnum)
                    foldernum = str(idnum) + '000'
                    mainpath = '\\\\' + 'rcstore.rc.fas.harvard.edu\geology\TMS Media\Archives\Accession and Object Records' + '\\' + (idnum)
                    if mainpath not in newpathdict.values(): 
                        newpathdict[idnum] = mainpath
                    finaldest = shutil.move(newpaths,mainpath)
                    
            if b[1] == '2':
                idnum = (b[1:8]).strip('.')  
                foldernum = (str(keys[:3])) + '-' + 'xxx'
                mainpath = '\\\\' + 'rcstore.rc.fas.harvard.edu\geology\TMS Media\Specimen\Accession and Object Records' + '\\' + (foldernum) 
                if mainpath not in newpathdict.values(): 
                    newpathdict[idnum] = mainpath
                finaldest = shutil.move(newpaths,mainpath)
                                     
        if 'minerals' in keys:
            i = keys.index('s\\')
            b = keys[i+1:]
            if b[1] == '1':
                    idnum = b[1:5]
                    foldernum = str(b[1:4]) + 'xxx'
                    mainpath = '\\\\'+'rcstore.rc.fas.harvard.edu\geology\TMS Media\Specimen\Minerals'+ '\\'+ (foldernum) + '\\' + (idnum) 
                    if mainpath not in newpathdict.values():
                        newpathdict[idnum] = mainpath
                    finaldest = shutil.move(keys,mainpath)
            if b[1] == '8' or b[1] == '9': 
                    idnum = b[1:6]
                    foldernum = '0' + str(b[1:3]) + 'xxx'
                    print(foldernum)
                    mainpath = '\\\\'+'rcstore.rc.fas.harvard.edu\geology\TMS Media\Specimen\Minerals'+ '\\'+ (foldernum) + '\\' + (idnum) 
                    newpathdict[idnum] = mainpath
                    if mainpath not in newpathdict.values():
                        newpathdict[idnum] = mainpath
                    finaldest = shutil.move(keys,mainpath)
                                                            
        if 'meteorites' in keys:
            i = keys.index('s\\')
            b = keys[i+1:]
            if b[1] == '2':
                idnum = (b[1:8]).strip('.')  
                foldernum = (str(b[1:5])) + '-' + 'xxx'
                mainpath = '\\\\' + 'rcstore.rc.fas.harvard.edu\geology\TMS Media\Specimen\Meteorites' + '\\' + (foldernum)
                newpathdict[foldernum] = mainpath
                if mainpath not in newpathdict.values(): 
                    newpathdict[idnum] = mainpath
                finaldest = shutil.move(keys,mainpath)
        if 'rocks' in newpaths:
            i = keys.index('s\\')
            b = keys[i+1:]
            if b[1] == '2':
                idnum = (b[1:8]).strip('.')  
                foldernum = (str(b[1:5])) + '-' + 'xxx'
                mainpath = '\\\\' + 'rcstore.rc.fas.harvard.edu\geology\TMS Media\Specimen\Rocks' + '\\' + (foldernum)
                if mainpath not in newpathdict.values():
                    newpathdict[idnum] = mainpath
                finaldest = shutil.move(keys,mainpath)
 
        if 'teaching' in keys:
            i = keys.index('n\\')
            b = keys[i+1:] 
            foldernum = str(b[:5]) + 'xxx'
            idnum = (b[:8])
            mainpath = '\\\\' + '\\rcstore.rc.fas.harvard.edu\geology\TMS Media\Specimen\Teaching Collection' + '\\' + (foldernum) + '\\' + (idnum)
            if mainpath not in newpathdict.values(): 
                newpathdict[idnum] = mainpath
            finaldest = shutil.move(keys,mainpath)

                            
                   
                
            with open('data.csv', 'w') as f:                                #creates a csv w filepaths and object IDs from newpaths dictionary
                f.write("%s, %s , %s\n" % ('filepath', 'obj number', 'obj number2'))
                for key in newpathdict.keys():
                    f.write("%s, %s\n" % (newpathdict[key], key))
                
          
                
        
```
